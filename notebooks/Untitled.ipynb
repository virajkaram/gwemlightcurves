```python
from gwemlightcurves.KNModels import KNTable
import pickle
import numpy as np
import glob
import pandas as pd
from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.table import vstack
```

```python

filt = 'J'
phi = 60
```

```python
samples = pd.read_csv('input/bns_samples_realistic.dat',header=0,delimiter='\t')
```

```python
crds = SkyCoord(ra=samples.iloc[180]['ra'],dec=samples.iloc[180]['dec'],unit=(u.rad,u.rad))
```

```python
ej_m = KNTable.read_samples('input/bns_samples_realistic_ejecta_masses.dat',Nsamples=1000)
```

```python
ej_crds = SkyCoord(ra=ej_m['ra'],dec=ej_m['dec'],unit=(u.rad,u.rad))
```

```python
idx,d2d,d3d = crds.match_to_catalog_sky(ej_crds)
```

```python
sel_src = ej_m[idx]
```

```python
sel_srcs = vstack([sel_src,sel_src,sel_src,sel_src,sel_src,sel_src,sel_src])
```

```python
sel_srcs['theta'] = [0,30,60,90,120,150,180]
```

```python
sel_srcs
```

```python
sel_srcs['tini'] = 0.01; sel_srcs['dt']=0.02;sel_srcs['tmax']=15
sel_srcs['phi'] = phi
```

```python
model_table = KNTable.model('Bu2019lm',sel_srcs,LoadModel=True,ModelPath='output/svdmodels/',filterlist=[filt])
```

```python
model_table
```

```python
pickle.dump(model_table,open('output/%sband_lcs_thetas_single_phi_%s'%(filt,phi),'wb'))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()
for row in model_table:
    plt.plot(row['t'],row['mag'][0],label=row['theta'])
plt.legend()
plt.gca().invert_yaxis()
```
